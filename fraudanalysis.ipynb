```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the transaction data dataset
transaction_data = pd.read_csv('data.csv')

# Calculate the proportion of fraudulent vs. non-fraudulent transactions
fraud_proportion = transaction_data['isFraud'].value_counts() / transaction_data['isFraud'].count()

# Get the number of fraudulent transactions of each type
n_fraudulent_transaction_per_type = transaction_data[transaction_data['isFraud'] == 1]['type'].value_counts()

# Create a pie chart to visualize the proportion of fraudulent vs. non-fraudulent transactions
plt.figure(1)
plt.pie(fraud_proportion, labels=['Fraudulent', 'Non-Fraudulent'], autopct='%1.1f%%')
plt.title('Proportion of Fraudulent vs. Non-Fraudulent Transactions')

# Plot a bar graph of the number of fraudulent transactions of each type
plt.figure(2)
plt.bar(n_fraudulent_transaction_per_type.index, n_fraudulent_transaction_per_type.values)
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Number of Fraudulent Transactions of Each Type')

# Identify the transaction type that's most commonly marked as fraudulent
most_commonly_fraudulent_transaction_type = n_fraudulent_transaction_per_type.idxmax()

# Print the most commonly fraudulent transaction type
print('The transaction type that\'s most commonly marked as fraudulent is', most_commonly_fraudulent_transaction_type)

plt.show()
```
