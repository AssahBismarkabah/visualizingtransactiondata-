```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the transaction data dataset
transaction_data = pd.read_csv('data.csv')

# Get the number of transactions of each type
n_transaction_per_type = transaction_data['type'].value_counts()

# Plot a bar graph of the number of transactions of each type
plt.bar(n_transaction_per_type.index, n_transaction_per_type.values)
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Number of Transactions of Each Type')
plt.show()

# Identify the most frequent transaction type
most_frequent_transaction_type = n_transaction_per_type.idxmax()

# Print the most frequent transaction type
print('The most frequent transaction type is', most_frequent_transaction_type)
```
